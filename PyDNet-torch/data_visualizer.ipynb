```python
import torch
import torchvision
import matplotlib.pyplot as plt
import numpy as np
from KittiDataset import KittiDataset
from Config import Config

# Funzione per mostrare un batch di immagini
def imshow(img):
    img = img / 2 + 0.5  # de-normalizza
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

config = Config('HomeLab').get_configuration()
dataset = KittiDataset(config.data_path,config.filenames_file_training,config.image_width,config.image_height,"train")
dataloader = dataset.make_dataloader(4,True)

dataiter = iter(dataloader)
left_batch, right_batch = dataiter.__next__()

imshow(torchvision.utils.make_grid(left_batch))
imshow(torchvision.utils.make_grid(right_batch))
```
